1. Graficar una distribución Normal con media 𝜇 = 10, y desviación estándar σ = 2




```python
miu = 10
sigma = 2

x = np.arange(miu - 4 * sigma, miu + 4 * sigma, 0.01)
y = norm.pdf(x, miu, sigma)

# Gráfica
plt.plot(x, y, color="red")
plt.title("Normal(0,1)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

2. Graficar una distribución T Student con grados de libertad 𝝂 = 12.

```python
gl = 12
sigma = np.sqrt(gl / (gl - 2))

x = np.arange(-4 * sigma, 4 * sigma, 0.01)
y = t.pdf(x, df=gl)

# Gráfica
plt.plot(x, y, color="blue")
plt.title("T Student with df = 12")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

3.  Gráfique la distribución Chi-cuadrada con 8 grados de libertad.

```python
gl = 8
sigma = np.sqrt(2*gl)

x = np.arange(0, miu + 8 * sigma, 0.01)
y = chi2.pdf(x, df=gl)

# Gráfica
plt.plot(x, y, color="green")
plt.title("T Student with df = 8")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```

5.  Si Z es una variable aleatoria que se distribuye normalmente con media 0 y desviación estándar 1, hallar los procedimientos de:

   a)   P(Z >0.7) = 0.2419637
   
   b)   P(Z < 0.7) = 0.7580363
   
   c)   P(Z = 0.7)  = 0

```python
z = 0.7

# Utilizando la función de distribución acumulativa
prob_z_b = norm.cdf(z, 0, 1)
prob_z_a = 1 - prob_z_b

# Utilizando la función de  densidad de probabilidad (PDF) de la distribución normal estándar
prob_z_c = norm.pdf(z, 0, 1)

print("P(Z > 0.7) =", prob_z_a)
print("P(Z < 0.7) =", prob_z_b)
print("P(Z = 0.7) =", prob_z_c)
```

6.  Cuando lo que se quiere es hallar el valor de Z dada el área a la izquierda bajo la curva se usa qnorm(área izq). Hallar el valor de Z que tiene al 45% de los demás valores inferiores a ese valor.

```python
percentile = 45
probability = percentile / 100  # Convertir el percentil a probabilidad acumulativa
z_value = norm.ppf(probability)

print("El valor Z para el percentil 45 es:", z_value)
```

7.  Hallar el procedimiento para verificar los siguientes resultados si se sabe que X se distribuye normalmente con una media de 100 y desviación estándar de 7.
 * P(X < 87) = 0.031645
 * P(X > 87) = 0.968354
 * P(87 < X < 110) = 0.89179

```python
miu = 100  # media
sigma = 7  # desviación estandar

x_value = 87

# Utilizando la función de distribución acumulativa
prob_a = norm.cdf(x_value, miu, sigma)
prob_b = 1- norm.cdf(x_value, miu, sigma)
prob_upper = norm.cdf(110, miu, sigma)
prob_c = prob_upper - prob_a

print("P(X < 87) =", prob_a)
print("P(X > 87) =", prob_b)
print("P(87 < X < 110) =", prob_c)
```

9. Hallar el procedimiento para verificar los siguientes resultados si se sabe que X se distribuye Chi-cuadrada con gl = 6, hallar
 * P(X2 < 3) = 0.1911532
 * P(X2 > 2) = 0.9196986
 * El valor x de chi que sólo el 5% de los demás valores de x es mayor a ese valor ( Resp. 12.59159)

```python
df = 6       # Grados de libertad
x_value_a = 3  # Valor de x inciso a
x_value_b = 2  # Valor de x inciso b

prob_a = chi2.cdf(x_value_a, df)
prob_b = 1 - chi2.cdf(x_value_b, df)

percentil = 0.95  # El 5% de los valores están por encima
x_value = chi2.ppf(percentil, df)

print("P(X^2 < 3) =", prob_a)
print("P(X^2 > 2) =", prob_b)
print("El valor de x para el percentil 5% es:", x_value)
```

10. Hallar el procedimiento para verificar los siguientes resultados si se sabe que X se distribuye F con v1 = 8, v2 = 10, hallar
 * P(X < 2) = 0.8492264
 * P(X > 3) = 0.05351256
 * El valor de x que sólo el 25% de los demás valores es inferior a él. (Resp. 0.6131229)

```python
v1 = 8  # Grados de libertad v1
v2 = 10  # Grados de libertad v2
x_value_a = 2  # Valor de x para el inciso a
x_value_b = 3  # Valor de x para el inciso b

prob_a = f.cdf(x_value_a, v1, v2)
prob_b = 1 - f.cdf(x_value_b, v1, v2)

percentile = 25 / 100  # Convertir el percentil a probabilidad
x_value_c = f.ppf(percentile, v1, v2)

print("P(X < 2) =", prob_a)
print("P(X > 3) =", prob_b)
print("El valor de x para el percentil 25% es:", x_value_c)
```
